```python
dataTrnsact.columns= ["SKU", "Store", "Register", "Trannum", "Interid", "Saledate", "Stype", "Quantity", "Unknown", "AMT", "Orgprice", "SEQ", "MIC", "Unknow"]
```

```python
DeptProduce = dataTrnsactFinal[dataTrnsactFinal['SKU'].isin(dataSKU[dataSKU['Dept'] == Dept]['SKU'])]
```

```python
DeptProduce.shape
```

```python
onehot= pd.get_dummies(DeptProduce['SKU'],prefix='SKU')
```

```python
df= pd.concat([DeptProduce,onehot],axis=1)
```

```python
df.head()
```

```python
df.drop(['SKU' , 'Store' , 'Interid' , 'Saledate' , 'Stype' , 'Quantity' , 'AMT' , 'Orgprice', 'MIC', 'Register', 'SEQ'], axis =1 , inplace = True)
```

```python
df.drop([ 'Unknown','Unknow'], axis = 1 , inplace = True)
```

```python
finalData = df.groupby(['Trannum']).sum()
```

```python
finalData.shape
```

```python
finalData[finalData>1]=1
```

```python
finalData.head()
```

```python
frequentItemsets = apriori(finalData , min_support=0.33, use_colnames=True)
```

```python
frequentItemsets
```

```python
frequentItemsets.shape
```

```python
rules = association_rules(frequentItemsets, metric="lift", min_threshold=1)
```

```python
rules.describe()
```

```python
rules[ rules['lift'] >= 2.5 ]
```

```python
rules[(rules['lift']>=2.36)&(rules['lift']<=2.37) & rules['confidence']>=0.9]
```
